```python
#Impact of Seed Individual, Random Population and Mutation

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots as sp

gen = 50

ci_df = pd.read_csv("test_out/counting_impact.csv")
#only_last_gen = data[data.generation == data.generation.max()]
only_last_gen_ci = ci_df[ci_df.generation == gen]
only_last_gen_ci['random_population'] = only_last_gen_ci[['traceID_2', 'traceID_3', 'traceID_4', 'traceID_5', 'traceID_6', 'traceID_7', 'traceID_8', 'traceID_9', 'traceID_10', 'traceID_11', 'traceID_12', 'traceID_13', 'traceID_14', 'traceID_15', 'traceID_16', 'traceID_17', 'traceID_18', 'traceID_19', 'traceID_20']].sum(axis=1)
only_last_gen_ci = only_last_gen_ci[['traceID_1', 'random_population', 'traceID_m', 'percentage_of_optimal_genes', 'seed_fitness', 'dim']]
#only_last_gen_ci = only_last_gen_ci[only_last_gen_ci.crossover_type == "combining"])

fig = sp.make_subplots(rows=3, cols=4, shared_yaxes=True, shared_xaxes=True)

fitnesses = [1,4,9,16]
dims = [4,8,12]


for col, fitness in enumerate(fitnesses):
    for row, dim in enumerate(dims):
        current_plot_df = only_last_gen_ci[only_last_gen_ci.seed_fitness ==fitness][only_last_gen_ci.dim == dim]
        fig.add_trace( go.Box(y=current_plot_df["traceID_1"], x=current_plot_df["percentage_of_optimal_genes"], name="seed ind", legendgroup="seed ind", showlegend=True), row=row+1, col=col+1)
        fig.add_trace( go.Box(y=current_plot_df["random_population"], x=current_plot_df["percentage_of_optimal_genes"], name="rest pop", legendgroup="rest pop", showlegend=True), row=row+1, col=col+1)
        fig.add_trace( go.Box(y=current_plot_df["traceID_m"], x=current_plot_df["percentage_of_optimal_genes"], name="mutation", legendgroup="mutation", showlegend=True), row=row+1, col=col+1)

fig.show()

#TODO: the percentage of optimal genes is not considered here!
```

```python
# Visualize Performance Data:
import pandas as pd
import plotly.express as px

performance_df = pd.read_csv("test_out/performance.csv")

performance_df = performance_df[performance_df.crossover_type == "combining"].drop(columns=["crossover_type"])
#performance_df = performance_df.drop(columns=["crossover_type"])

mean_performance_df = performance_df.groupby(["percentage_of_optimal_genes", "seed_fitness", "generation", "dim"], as_index=False).mean()


fig = px.line(mean_performance_df[mean_performance_df.dim ==4], y="min_f", x="generation", color="seed_fitness", symbol="percentage_of_optimal_genes")
fig.show()
fig.update_layout(yaxis_type="log")
fig.show()

fig = px.line(mean_performance_df[mean_performance_df.dim ==8], y="min_f", x="generation", color="seed_fitness", symbol="percentage_of_optimal_genes")
fig.show()
fig.update_layout(yaxis_type="log")
fig.show()

fig = px.line(mean_performance_df[mean_performance_df.dim ==12], y="min_f", x="generation", color="seed_fitness", symbol="percentage_of_optimal_genes")
fig.show()
fig.update_layout(yaxis_type="log")
fig.show()
```
