```python
# Visualize Performance Data:
import pandas as pd
import plotly.express as px

performance_df = pd.read_csv("test_out/performance.csv")
performance_df = performance_df[performance_df.crossover_type == "combine"].drop(columns=["crossover_type"])

mean_performance_df = performance_df.groupby(["seed_type", "seed_fitness", "generation"], as_index=False).mean()

fig = px.line(mean_performance_df, y="min_f", x="generation", color="seed_fitness", symbol="seed_type")
fig.show()
fig.update_layout(yaxis_type="log")
fig.show()
```
